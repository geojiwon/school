```python
def printStep(arr, val, method, is_original=False):
    if is_original:
        print(f"Original      : {arr}")
    elif method:
        print(f"Step {val}        : {arr}")
    else:
        print(f"Step {val}          : {arr}")


def selection_sort(A):
    n = len(A)
    for i in range(n - 1):
        least = i
        for j in range(i + 1, n):
            if A[j] < A[least]:
                least = j
        A[i], A[least] = A[least], A[i]
        printStep(A, i + 1, "Selection")


def insertion_sort(A):
    n = len(A)
    for i in range(1, n):
        key = A[i]
        j = i - 1
        while j >= 0 and key < A[j]:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = key
        printStep(A, i, "Insertion")


def bubble_sort(A):
    n = len(A)
    for i in range(n - 1, 0, -1):
        bChanged = False
        for j in range(i):
            if A[j] > A[j + 1]:
                A[j], A[j + 1] = A[j + 1], A[j]
                bChanged = True
        if not bChanged:
            break
        printStep(A, n - i, "Bubble")


while True:
    print("1. 선택 정렬")
    print("2. 삽입 정렬")
    print("3. 버블 정렬")
    print("4. 종료")

    choice = input("정렬 방법을 선택하세요 (1-4): ")

    if choice == '1':
        A = [5, 3, 8, 4, 9, 1, 6, 2, 7]
        printStep(A, None, None, is_original=True)
        selection_sort(A)
        print(f"Selection   : {A}")

    elif choice == '2':
        A = [5, 3, 8, 4, 9, 1, 6, 2, 7]
        printStep(A, None, None, is_original=True)
        insertion_sort(A)
        print(f"Insertion   : {A}")

    elif choice == '3':
        A = [5, 3, 8, 4, 9, 1, 6, 2, 7]
        printStep(A, None, None, is_original=True)
        bubble_sort(A)
        print(f"Bubble   : {A}")

    elif choice == '4':
        print("프로그램을 종료합니다.")
        break

    else:
        print("올바른 선택이 아닙니다. 1-4 중에서 선택하세요.")
```

```python
class Entry:
    def __init__(self, key, value):
        self.key = key
        self.value = value

    def __str__(self):
        return str("%s:%s" % (self.key, self.value))


def print_entries(entries):
    for entry in entries:
        print(entry, end=' ')


def sequential_search(A, key, low, high):
    for i in range(low, high + 1):
        if A[i].key == key:
            return i
    return None


def binary_search(A, key, low, high):
    if low <= high:
        middle = (low + high) // 2
        if key == A[middle].key:
            return middle
        elif key < A[middle].key:
            return binary_search(A, key, low, middle - 1)
        else:
            return binary_search(A, key, middle + 1, high)
    return None


arr = [Entry(2, 'a'), Entry(6, 'b'), Entry(11, 'c'), Entry(13, 'd'),
       Entry(18, 'e'), Entry(20, 'f'), Entry(22, 'g'), Entry(27, 'h'),
       Entry(29, 'i'), Entry(30, 'j'), Entry(34, 'k'), Entry(38, 'l'),
       Entry(41, 'm'), Entry(42, 'n'), Entry(45, 'o'), Entry(47, 'p')]

while True:

    choice = input("1.순차탐색  2. 이진탐색  3.종료  : ")

    if choice == '1':
        print("순차탐색을 선택하셨습니다.")
        print_entries(arr)
        key = int(input("\n찾을 키 값을 입력하세요: "))
        result = sequential_search(arr, key, 0, len(arr) - 1)
        if result is not None:
            print(f"key = {key} : value = {arr[result].value}")
        else:
            print(f"key = {key}를 찾을 수 없습니다.")
        print("")

    elif choice == '2':
        print("이진탐색을 선택하셨습니다.")
        print_entries(arr)
        key = int(input("\n찾을 키 값을 입력하세요: "))
        result = binary_search(arr, key, 0, len(arr) - 1)
        if result is not None:
            print(f"key = {key} : value = {arr[result].value}")
        else:
            print(f"key = {key}를 찾을 수 없습니다.")
        print("")

    elif choice == '3':
        print("프로그램을 종료합니다.")
        break

    else:
        print("잘못된 선택입니다. 1-3 중에서 선택하세요.")
        print("")
```
